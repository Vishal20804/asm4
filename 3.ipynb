```python
Ans:4
The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials, where each trial has the same probability of success. The properties of the binomial distribution include:

    fixed numbers
    independent trials
    constant probability
    discreate distribution
```

```python
Ans:5
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters
n = 1000  # Number of trials
p = 0.4  # Probability of success

# Generate the random sample
sample = np.random.binomial(n, p, size=1000)

# Plot the histogram
plt.hist(sample, bins='auto', alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Histogram of Random Sample from Binomial Distribution')
plt.show()
```

```python
Ans:6
import math

def poisson_cdf(k, mean):
    cdf = 0
    for i in range(k + 1):
        cdf += math.exp(-mean) * (mean ** i) / math.factorial(i)
    return cdf
mean = 2.5  # Mean of the Poisson distribution
k = 4  # Point at which you want to calculate the CDF

cdf = poisson_cdf(k, mean)
print(f"The CDF at k = {k} is {cdf:.4f}")
```

```python
Ans:7
The binomial distribution and the Poisson distribution are both probability distributions that are used to model events, but they differ in several key aspects:

Nature of Events:

Binomial Distribution: The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials, where each trial has the same probability of success.
Poisson Distribution: The Poisson distribution models the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence and independence between events.
Type of Trials:

Binomial Distribution: The trials in a binomial distribution are of a fixed size and can result in a success or failure. The number of trials, denoted as 'n', is predetermined.
Poisson Distribution: The trials in a Poisson distribution are not predetermined in terms of size. Instead, the distribution models the occurrence of events over a continuous interval, such as time or space.
```
